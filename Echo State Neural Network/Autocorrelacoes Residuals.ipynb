```python
import pandas
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
target = "Dogecoin"
best = pandas.read_csv("BestPredictor" + target + ".txt", header = None).to_numpy()[:,0]
worst = pandas.read_csv("WorstPredictor" + target + ".txt", header = None).to_numpy()[:,0]
best.shape
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14, 8),facecolor='white')
plot_acf(best, lags = 25, title=target + 's Residual Autocorrelation (Best Model)', ax=ax[0, 0], zero = False, color = "red", vlines_kwargs={"colors": "red"})
plot_pacf(best, lags = 25, title=target + 's Residual Partial Autocorrelation (Best Model)', ax=ax[0, 1], zero = False, color = "red", vlines_kwargs={"colors": "red"})
plot_acf(worst, lags = 25, title=target + 's Residual Autocorrelation (Worst Model)', ax=ax[1, 0], zero = False, color = "red", vlines_kwargs={"colors": "red"})
plot_pacf(worst, lags = 25, title=target + 's Residual Partial Autocorrelation (Worst Model)', ax=ax[1, 1], zero = False, color = "red", vlines_kwargs={"colors": "red"})
plt.savefig("Autocorrelation" + target + "Residuals.png", facecolor=fig.get_facecolor(), transparent = True, dpi=300)
plt.show()
```
