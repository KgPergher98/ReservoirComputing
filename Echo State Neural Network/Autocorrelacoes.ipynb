```python
import pandas
import statsmodels.api as sm
import matplotlib.pyplot as plt
import ESN_Modules as ESN
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.preprocessing import PowerTransformer
```

```python
x1 = pandas.read_csv("BitcoinData.txt", header = None).to_numpy()[:,0]
x2 = pandas.read_csv("EthereumData.txt", header = None).to_numpy()[:,0]
x3 = pandas.read_csv("CardanoData.txt", header = None).to_numpy()[:,0]
x4 = pandas.read_csv("DogecoinData.txt", header = None).to_numpy()[:,0]

y = x1.shape[0]
x1 = PowerTransformer().fit_transform(ESN.reshape(x1[0:int(0.7*y)]))
x2 = PowerTransformer().fit_transform(ESN.reshape(x2[0:int(0.7*y)]))
x3 = PowerTransformer().fit_transform(ESN.reshape(x3[0:int(0.7*y)]))
x4 = PowerTransformer().fit_transform(ESN.reshape(x4[0:int(0.7*y)]))
x1.shape
```

```python
name = "Dogecoin"
data = x4
fig, ax = plt.subplots(1, 2, figsize=(12, 4),facecolor='white')
plot_acf(data, lags = 25, title=name + ' Autocorrelation', ax=ax[0], zero = False)
plot_pacf(data, lags = 25, title=name + ' Partial Autocorrelation', ax=ax[1], zero = False)
plt.savefig("Autocorrelation" + name + ".png", facecolor=fig.get_facecolor(), transparent = True, dpi=300)
plt.show()
```
