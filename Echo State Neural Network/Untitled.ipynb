```python
import matplotlib.pyplot as plt
import pandas

target = "Bitcoin"
df = pandas.read_csv(target + "Data.txt", header = None)
df2 = pandas.read_csv(target + "TrainData.txt", header = None)

#print(df.describe())
#print(df.skew())
#print(df.kurtosis())
#print("Norm")
#print(df2.describe())
#print(df2.skew())
#print(df2.kurtosis())

fig, ax = plt.subplots(1, 2, figsize=(10, 4),facecolor='white')
ax[0].hist(df[0].to_numpy(), bins = 200, color = "blue", label = "Log-Returns")
ax[1].hist(df2[0].to_numpy(), bins = 200, color = "red", label = "Normalised Log-Returns")
ax[0].legend()
ax[1].legend()
plt.suptitle('Distribuções de retorno: ' + target, fontsize=14)
plt.savefig("Distribution" + target + ".png", dpi=300)
plt.show()
```
